```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

def create_architecture_diagram():
    # Setup Canvas
    fig, ax = plt.subplots(figsize=(14, 9))
    ax.set_xlim(0, 14)
    ax.set_ylim(0, 9)
    ax.axis('off')

    # Helper function to draw boxes
    def draw_node(x, y, w, h, label, color, type_label=""):
        # Shadow
        shadow = patches.FancyBboxPatch((x+0.1, y-0.1), w, h, boxstyle="round,pad=0.1", 
                                        linewidth=0, facecolor='gray', alpha=0.3)
        ax.add_patch(shadow)
        # Box
        box = patches.FancyBboxPatch((x, y), w, h, boxstyle="round,pad=0.1", 
                                     linewidth=2, edgecolor='#333', facecolor=color)
        ax.add_patch(box)
        # Text
        ax.text(x + w/2, y + h/2, label, ha='center', va='center', fontsize=11, fontweight='bold', wrap=True)
        if type_label:
            ax.text(x + w/2, y + h + 0.2, type_label, ha='center', va='bottom', fontsize=9, fontstyle='italic', color='#555')
        return (x + w/2, y, x + w/2, y + h) # Bottom-Mid, Top-Mid

    # --- Title ---
    ax.text(7, 8.5, "System Architecture: AI Concrete Engineering Suite", ha='center', fontsize=18, fontweight='bold', color='#222')

    # --- ZONES ---
    # User Interface Zone
    ax.add_patch(patches.Rectangle((0.5, 0.5), 13, 2.5, linewidth=1, edgecolor='#ddd', facecolor='#f9f9f9', linestyle='--'))
    ax.text(1, 2.7, "USER INTERFACE (Gradio)", fontsize=10, fontweight='bold', color='#777')

    # Processing Zone
    ax.add_patch(patches.Rectangle((0.5, 3.5), 13, 4.5, linewidth=1, edgecolor='#ddd', facecolor='#f0f8ff', linestyle='--'))
    ax.text(1, 7.7, "BACKEND LOGIC & AI MODELS", fontsize=10, fontweight='bold', color='#777')

    # --- NODES ---
    
    # 1. Inputs
    draw_node(1.5, 1.0, 3, 1, "Mix Parameters\n(Cement, Water, etc.)", "#FFD700", "Tab Input")
    draw_node(9.5, 1.0, 3, 1, "Site Image\n(Upload)", "#FFD700", "Image Input")

    # 2. Preprocessing
    draw_node(1.5, 4.0, 3, 1, "Data Scaling\n(StandardScaler)", "#ADD8E6", "Preprocessing")
    draw_node(9.5, 4.0, 3, 1, "Image Resizing\n(128x128 px)", "#ADD8E6", "Preprocessing")

    # 3. Models
    draw_node(1.5, 6.0, 3, 1, "Strength Prediction\n(ANN Regression)", "#98FB98", "Model 1")
    draw_node(9.5, 6.0, 3, 1, "Crack Detection\n(CNN Classification)", "#98FB98", "Model 2")

    # 4. Output Logic
    draw_node(5.5, 2.0, 3, 1, "Result Display\n(Prediction & Confidence)", "#FFA07A", "Output")

    # --- ARROWS ---
    arrow_props = dict(facecolor='black', edgecolor='black', arrowstyle='simple', alpha=0.6)

    # Path 1: Strength
    ax.annotate("", xy=(3, 4), xytext=(3, 2.2), arrowprops=arrow_props) # Input -> Scale
    ax.annotate("", xy=(3, 6), xytext=(3, 5.2), arrowprops=arrow_props) # Scale -> Model
    ax.annotate("", xy=(5.4, 2.5), xytext=(3, 6.5), arrowprops=dict(connectionstyle="arc3,rad=-0.2", **arrow_props)) # Model -> Output

    # Path 2: Cracks
    ax.annotate("", xy=(11, 4), xytext=(11, 2.2), arrowprops=arrow_props) # Input -> Resize
    ax.annotate("", xy=(11, 6), xytext=(11, 5.2), arrowprops=arrow_props) # Resize -> Model
    ax.annotate("", xy=(8.6, 2.5), xytext=(11, 6.5), arrowprops=dict(connectionstyle="arc3,rad=0.2", **arrow_props)) # Model -> Output

    plt.tight_layout()
    plt.savefig('architecture.png')
    print("Architecture diagram saved as 'architecture.png'")
    plt.show()

create_architecture_diagram()
```
